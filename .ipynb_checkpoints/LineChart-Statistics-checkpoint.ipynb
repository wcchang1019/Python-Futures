```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('LineChart_onehour.csv')
```

```python
df = df.drop(df.index[0])
```

```python
df = df.reset_index()
```

```python
del df['index']
```

```python
dfgroup = df.groupby(by='Category')
dfgroupedlist = list(dfgroup)
```

```python
#[x for x in dfgroup]
dfgroupedlist[0][1]['All_Time']
```

```python
CategorySize = list(dfgroup.size())
labels = ['b-horn','b-wait','s-wait','s-convergence','s-horn','b-strong','s-strong','buy-convergence','b-end','s-end']
colors = ['blue','red','green','yellow','blue','red','green','yellow','red','green']
```

```python
plt.pie(CategorySize , labels=labels , colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
```

```python
fig , (ax0,ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9) = plt.subplots(figsize=(16,20),nrows=10)
plt.subplots_adjust(hspace = 0.4)
ax0.hist(dfgroupedlist[0][1]['All_Time'],bins=30,edgecolor='black')
ax0.set_title('b-horn')
ax1.hist(dfgroupedlist[1][1]['All_Time'],bins=30,edgecolor='black')
ax1.set_title('b-wait')
ax2.hist(dfgroupedlist[2][1]['All_Time'],bins=30,edgecolor='black')
ax2.set_title('s-wait')
ax3.hist(dfgroupedlist[3][1]['All_Time'],bins=30,edgecolor='black')
ax3.set_title('s-convergence')
ax4.hist(dfgroupedlist[4][1]['All_Time'],bins=30,edgecolor='black')
ax4.set_title('s-horn')
ax5.hist(dfgroupedlist[5][1]['All_Time'],bins=30,edgecolor='black')
ax5.set_title('b-strong')
ax6.hist(dfgroupedlist[6][1]['All_Time'],bins=30,edgecolor='black')
ax6.set_title('s-strong')
ax7.hist(dfgroupedlist[7][1]['All_Time'],bins=30,edgecolor='black')
ax7.set_title('b-convergence')
ax8.hist(dfgroupedlist[8][1]['All_Time'],bins=30,edgecolor='black')
ax8.set_title('b-end')
ax9.hist(dfgroupedlist[9][1]['All_Time'],bins=30,edgecolor='black')
ax9.set_title('s-end')
```

```python
pre = list(df['Category'])
pre.insert(0,0)
pre.pop()
```

```python
df['pre'] = pre
```

```python
df_group_pre = df.groupby(['pre','Category'])
```

```python
#[x for x in df_group_pre]
df_group_pre.size()
```

```python
dict_category = {1:'b-horn',2:'b-wait',3:'s-wait',4:'s-convergence',5:'s-horn',6:'b-strong',7:'s-strong',8:'buy-convergence',9:'b-end',10:'s-end'}
```

```python
def plot_know_pre_pie(num):
    piesize = list(df_group_pre.size()[num])
    labels = list(map(lambda x :dict_category[num],list(df_group_pre.size()[num].index)))
    plt.figure(figsize=(6,6))
    plt.pie(piesize,labels=labels,autopct='%1.1f%%')
    plt.title('pre is '+dict_category[num])
```

```python
plot_know_pre_pie(1)
plot_know_pre_pie(2)
plot_know_pre_pie(3)
plot_know_pre_pie(4)
plot_know_pre_pie(5)
plot_know_pre_pie(6)
plot_know_pre_pie(7)
plot_know_pre_pie(8)
plot_know_pre_pie(9)
plot_know_pre_pie(10)
```

```python
axes = dfgroupedlist[0][1].boxplot(column = 'Amp_3',by='Time_3',fontsize=10)
plt.title('')
fig = axes.get_figure()
fig.suptitle(dict_category[1], fontsize=14, fontweight='bold')
```

```python
def plot_box_Amp3_Time3(num):
    axes = dfgroupedlist[num-1][1].boxplot(column = 'Amp_3',by='Time_3',fontsize=10)
    plt.title('')
    fig = axes.get_figure()
    fig.suptitle(dict_category[num], fontsize=14, fontweight='bold')
```

```python
plot_box_Amp3_Time3(1)
plot_box_Amp3_Time3(2)
plot_box_Amp3_Time3(3)
plot_box_Amp3_Time3(4)
plot_box_Amp3_Time3(5)
plot_box_Amp3_Time3(6)
plot_box_Amp3_Time3(7)
plot_box_Amp3_Time3(8)
plot_box_Amp3_Time3(9)
plot_box_Amp3_Time3(10)
```
